```python
import tensorflow as tf
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')


x_train_path = "/content/drive/MyDrive/Weights/x_train.xlsx";
y_train_path = "/content/drive/MyDrive/Weights/y_train.xlsx";
x_test_path = "/content/drive/MyDrive/Weights/x_test.xlsx";
y_test_path = "/content/drive/MyDrive/Weights/y_test.xlsx";

x_train_pd = pd.read_excel(x_train_path)
y_train_pd = pd.read_excel(y_train_path)

x_test_pd = pd.read_excel(x_test_path)
y_test_pd = pd.read_excel(y_test_path)

x_train = x_train_pd.values
y_train = y_train_pd.values

x_test = x_test_pd.values
y_test = y_test_pd.values


Input = tf.keras.layers.Input(shape=(250, 1), name='Input_Layer');
Hidden_1 = tf.keras.layers.LSTM(5, return_sequences=True, activation='tanh', name='Hidden_Layer_1')(Input);
Hidden_2 = tf.keras.layers.LSTM(5, return_sequences=False, activation='tanh', name='Hidden_Layer_2')(Hidden_1);
Output = tf.keras.layers.Dense(4, activation= 'softmax', name= "Output_Layer")(Hidden_2);
model = tf.keras.Model(inputs=Input, outputs=Output);

model = tf.keras.models.load_model('/content/drive/MyDrive/Saved_Model_10x10/model_2.h5')
tf.keras.models.save_model(model, '/content/drive/MyDrive/Saved_Model_10x10/model_3_NT.h5')
model_nt = tf.keras.models.load_model('/content/drive/MyDrive/Saved_Model_10x10/model_3_NT.h5')

model.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = False),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001),
    metrics = ['accuracy']
)

model.fit(x_train, y_train, batch_size=1, epochs=10, verbose=1)
model.evaluate(x_test, y_test, batch_size=1, verbose=1)


predictions = model.predict(x_test)
print(y_test[14])
print(predictions[14])

print(model.trainable_weights)

for layer in model.layers:
    weights, biases = layer.get_weights()
    print(f"Layer: {layer.name}")
    print(f"  Weights:\n{weights}")
    print(f"  Biases:\n{biases}")
    print()
```
